```python
train = pd.read_csv('./train.csv')
val = pd.read_csv('./val.csv')
test = pd.read_csv('./test.csv')
sample_submission = pd.read_csv('./sample_submission.csv')
```

# 데이터 표준화

## 검증 데이터

```python
from sklearn.preprocessing import StandardScaler  # 표준화 
val_x = val.drop(['Class'], axis=1).values # 독립변인들의 value값만 추출
val_y = val['Class'].values # 종속변인 추출
```

```python
std = StandardScaler()
val_x = std.fit_transform(val_x) # 독립변인 표준화
```

## 학습 데이터

```python
train_stdscale = std.fit_transform(train)
```

# PCA 실시

## 검증 데이터

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=25) # 주성분 개수
printcipalComponents = pca.fit_transform(val_x)
pca_valid = pd.DataFrame(data=printcipalComponents)
pca_valid['Class'] = val_y
pca_valid.head(30)
```

```python
print(pca.explained_variance_ratio_)
print(sum(pca.explained_variance_ratio_))
```

## 학습 데이터

```python
pca = PCA(n_components=25) # 주성분 개수
printcipalComponents = pca.fit_transform(train_stdscale)
pca_train = pd.DataFrame(data=printcipalComponents)
pca_train.head(30)
```

```python
print(pca.explained_variance_ratio_)
print(sum(pca.explained_variance_ratio_))
```
